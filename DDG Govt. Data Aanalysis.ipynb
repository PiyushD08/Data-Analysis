```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('List_of_Ayurveda_Hospitals_District_wise.csv')
df
```

```python
#Checking missing values
df.isna().sum()
```

```python
# remove duplicate rows
print(df.drop_duplicates())
print('\n')
print('Number of duplicate rows are:')
print(df.duplicated().sum())
```

```python
# remove unnecessary columns
df = df.drop(['Taluk '], axis=1)
df
```

```python
# Calculate summary statistics
mean_bed_strength = df["Bed Strength"].mean()
median_bed_strength = df["Bed Strength"].median()
mode_bed_strength = df["Bed Strength"].mode().values[0]
std_bed_strength = df["Bed Strength"].std()

print("Mean Bed Strength:", mean_bed_strength)
print("Median Bed Strength:", median_bed_strength)
print("Mode Bed Strength:", mode_bed_strength)
print("Standard Deviation of Bed Strength:", std_bed_strength)
```

```python
#Data Description
df.describe
```

```python
#Bar Chart of the number of hospitals in each district.
df.groupby("District")["Types "].count().plot(kind='bar',figsize=(15,5))
plt.title("Number of Hospitals by District")
plt.xlabel("District name")
plt.ylabel("Number of Hospitals")
plt.show()
```

```python
# create a new column 'Location' based on the values of 'Rural/Urban'
df['Location'] = df['Rural/Urban'].apply(lambda x: 'Rural' if x == 1 else 'Urban')

# create a box plot of bed strength by location
df.boxplot(column='Bed Strength', by='Location', figsize=(5,5))
plt.title("Bed strength by Location (Rural/Urban)")
plt.xlabel("Location")
plt.ylabel("Bed strength")
plt.show()
```

```python
# create a pivot table with District as rows and Bed Strength as columns
pivot_table = df.pivot_table(values="Bed Strength", index="District", aggfunc="mean")

# create the heatmap
sns.heatmap(pivot_table, cmap="YlGnBu")
plt.show()
```

```python
# Plot the scatter graph
plt.scatter(df['Bed Strength'], df['Rural/Urban'])

# Add labels and title to the graph
plt.xlabel("Bed Strength")
plt.ylabel("Location")
plt.title("Bed Strength vs Location Scatter Plot")

# Show the graph
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#linear regression model to predict the bed strength of a hospital based on type of hospitals 
X = df[["District", "Types "]]
y = df["Bed Strength"]

# Converting categorical variables to dummy variables
X = pd.get_dummies(X)

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

# Creating a linear regression model
model = LinearRegression()

# Fitting the model to the training data
model.fit(X_train, y_train)

# Making predictions on the test data
y_pred = model.predict(X_test)

# Calculating the mean squared error of the predictions
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", mse)
```
